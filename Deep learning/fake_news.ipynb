## Fake news classifier

### data loading

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../Final.csv')
df.head()
```

```python
df = df.sample(frac = 1).reset_index(drop = True)
df.head()
```

```python
df.drop('Unnamed: 0', axis = 1, inplace = True)
df.head()
```

```python
df
```

```python
df['info'].value_counts()
```

```python
X = df.drop('info', axis = 1)
y = df['info']
```

### preprocessing the text

```python
messages = X.copy()
```

```python
import nltk
import re
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
corpus = []
ps = PorterStemmer()
for i in range(0, len(messages)) : 
    review = re.sub('^[a-zA-Z]', ' ', messages['title'][i])

    review = review.lower()
    review = review.split()

    review = [ps.stem(words) for words in review if not words in stopwords.words('english')]

    review = ' '.join(review)

    corpus.append(review)
```

```python
corpus
```

```python
X_final = np.array(corpus)
y_final = y
```

```python
X_final.shape
```

```python
y_final.shape
```

### word embedding

```python
import tensorflow
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
vocab_size = 10000
one_hot_rep = [one_hot(word, vocab_size) for word in corpus]
```

```python
one_hot_rep = pad_sequences(one_hot_rep, maxlen = 30, padding = 'pre')
```

```python
one_hot_rep
```

```python
X_final = np.array(one_hot_rep)
```

```python
X_final.shape
```

```python
y_final.shape
```

### train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size = 0.25, random_state = 42)
```

```python
print(f"{X_train.shape} :: {X_test.shape} :: {y_train.shape} :: {y_test.shape}")
```

### model building and training

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Bidirectional, LSTM, Embedding, Dense, Dropout
```

```python
model = Sequential([
    Embedding(input_dim = 10000, output_dim = 30, input_length = 30),
    Bidirectional(LSTM(64, return_sequences = True)),
    Dropout(0.3),
    Bidirectional(LSTM(32)),
    Dense(64, activation = 'relu'),
    Dense(1, activation = 'sigmoid')
])
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
model.summary()
```

### model training and predictions

```python
model.fit(X_train, y_train, epochs = 10, batch_size = 128, validation_split = 0.20)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred = (y_pred > 0.5).astype(int)
```

```python
y_pred
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
print(accuracy_score(y_pred, y_test))
print(confusion_matrix(y_pred, y_test))
print(classification_report(y_pred, y_test))
```
